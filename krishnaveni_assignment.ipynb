Area of a Triangle (Using Heron‚Äôs Formula)

Enter side a: 5

Enter side c: 7
Sample Output:

```python
a=int(input("enter the side a"))
b=int(input("enter the side b"))
c=int(input("enter the side c"))
s=(a+b+c)/2
area=(s*(s-a)*(s-b)*(s-c))**0.5
print("area of the triangle",area)
```
